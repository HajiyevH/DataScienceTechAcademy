```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msn
df  = pd.read_csv("/Users/hajiaga/Downloads/archive/athlete_events.csv")
```

```python
numerics = df._get_numeric_data().columns
def ikililer(data , column):
    fig ,rows = plt.subplots(figsize = [12 , 4] ,ncols = 2 , nrows = 1)
    sns.boxplot(ax = rows[0] , data = df , x = column)
    plt.title(column)
    sns.histplot(ax = rows[1], data = df , x = column)
for col in numerics:
    ikililer(df, col)
numerics
```

```python
msn.bar(df)
plt.show()
```

```python
Sport_names = list(df["Sport"].unique())

for i in Sport_names:
    if df[df["Sport"] == i].shape[0] > 100:
        sns.scatterplot(x = "Height" , y = "Weight" , data = df[df["Sport"] == i])
        plt.title(f"Correlation between Height and Weight in {i}")
        plt.xlabel("Height")
        plt.ylabel("Weight")
        plt.legend()
        plt.show()

```

```python
res = df.groupby(["Team" , "Medal"]).size().reset_index()
res.columns=["Country" , "Medal" , "Values"]
res = res.pivot(index="Country" , columns="Medal" ,values= "Values" )
res.fillna(0)
res.sort_values("Gold" , ascending=False)[:20].plot(kind="bar")
```

```python
nese = df.groupby(["Year" , "Sex"]).size().reset_index()
nese.replace("M" , "Male" , inplace=True)
nese.replace("F" , "Female" , inplace=True)

nese.columns = ["Year" , "Gender" , "Count"]
nese = nese.pivot(index = "Year" , columns= "Gender" , values="Count")
nese.fillna(0)
nese.sort_values(by = "Male")
nese.plot(kind = "bar")
plt.show()
```
